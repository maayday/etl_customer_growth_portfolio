```python
import pandas as pd

# Load your CSV
df = pd.read_csv("../data/customer_growth_2025-04-24.csv")  # <-- update date if needed
df.head()
```

```python
import psycopg2
from db import get_connection

# Reconnect to the DB
conn = get_connection()
cur = conn.cursor()

# Create the table if it doesn't exist
cur.execute("""
CREATE TABLE IF NOT EXISTS monthly_customer_growth (
    join_month VARCHAR(7),
    new_customers INT
);
""")
conn.commit()

# Insert each row from the CSV into the table
for _, row in df.iterrows():
    cur.execute("""
        INSERT INTO monthly_customer_growth (join_month, new_customers)
        VALUES (%s, %s)
        ON CONFLICT DO NOTHING;
    """, (row["join_month"], row["new_customers"]))

conn.commit()
cur.close()
conn.close()

print("✅ Data loaded into PostgreSQL!")
```

```python
conn = get_connection()
cur = conn.cursor()

cur.execute("SELECT * FROM monthly_customer_growth ORDER BY join_month")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```
