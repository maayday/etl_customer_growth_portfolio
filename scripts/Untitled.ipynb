```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/customer_growth_2025-04-24.csv")  # Update filename as needed
```

```python
df["join_month"] = pd.to_datetime(df["join_month"])
df = df.sort_values("join_month")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["join_month"], df["new_customers"], marker='o', linewidth=2)
plt.title("Monthly Customer Growth", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("New Customers", fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
